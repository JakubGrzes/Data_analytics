Task 2.

```python
uniform_distribution = np.random.uniform(low=0.0, high=1.0, size=1000)
normal_distribution_sigma_1 = np.random.normal(0, 1, 1000)
normal_distribution_sigma_10 = np.random.normal(0, 10, 1000)
beta_distribution_1_1 = np.random.beta(1, 1, 1000)
beta_distribution_2_2 = np.random.beta(2, 2, 1000)
beta_distribution_3_9 = np.random.beta(3, 9, 1000)
normal_2D_distribution_1 = np.random.multivariate_normal([0, 0], [[1, 0], [0, 1]], 10000)
normal_2D_distribution_2 = np.random.multivariate_normal([0, 0], [[2, 1], [1, 2]], 10000)
```

Task 3.

```python
def plot_1D_hist(samples, title, distribution, parameters) -> None:
  plt.figure()
  plt.hist(samples, 100, density=True)
  plt.title(title)
  plt.xlabel('Value')
  plt.ylabel('Amount of samples')
  plt.grid(True)
  x = np.arange(min(samples), max(samples), (max(samples) - min(samples)) / 100)
  if distribution == 'uniform':
    y = stats.uniform.pdf(x)
  elif distribution == 'normal':
    y = stats.norm.pdf(x, parameters[0], parameters[1])
  elif distribution == 'beta':
    y = stats.beta.pdf(x, parameters[0], parameters[1])
  else:
    print(f'Distribution "{distribution}" unknown')
  plt.plot(x, y)
  plt.show()

plot_1D_hist(uniform_distribution, title='uniform distribution histogram', distribution='uniform', parameters=[])
plot_1D_hist(normal_distribution_sigma_1, title='normal distribution histogram: mu=0, sigma=1', distribution='normal', parameters=[0, 1])
plot_1D_hist(normal_distribution_sigma_10, title='normal distribution histogram: mu=0, sigma=10', distribution='normal', parameters=[0, 10])
plot_1D_hist(beta_distribution_1_1, title='Beta distribution histogram: alpha=1, Beta=1', distribution='beta', parameters=[1, 1])
plot_1D_hist(beta_distribution_2_2, title='Beta distribution histogram: alpha=2, Beta=2', distribution='beta', parameters=[2, 2])
plot_1D_hist(beta_distribution_3_9, title='Beta distribution histogram: alpha=3, Beta=9', distribution='beta', parameters=[3, 9])
```

Taks 4.

```python
def plot_2D_hist(samples, title, bins) -> None:
  plt.figure()
  plt.hist2d(samples[:,0], samples[:,1], bins)
  plt.title("matplotlib, " + title + " , bins=" + str(bins))
  plt.xlabel('x value')
  plt.ylabel('y value')
  plt.grid(True)
  plt.show()

  sns.histplot(x=samples[:,0], y=samples[:,1], bins=bins)
  plt.title("seaborn, " + title + " , bins=" + str(bins))
  plt.xlabel('x value')
  plt.ylabel('y value')
  plt.grid(True)

plot_2D_hist(normal_2D_distribution_1, title='normal 2D distribution: mu=[0,0].T, cov=[[1,0], [0,1]]', bins=10)
plot_2D_hist(normal_2D_distribution_1, title='normal 2D distribution: mu=[0,0].T, cov=[[1,0], [0,1]]', bins=100)
plot_2D_hist(normal_2D_distribution_2, title='normal 2D distribution: mu=[0,0].T, cov=[[2,1], [1,2]]', bins=10)
plot_2D_hist(normal_2D_distribution_2, title='normal 2D distribution: mu=[0,0].T, cov=[[2,1], [1,2]]', bins=100)
```

```python
alpha = np.random.normal(2, 0.1, 1000)
beta = np.random.normal(0.2, 0.2, 1000)
x = np.random.normal(0, 3, 1000)
sigma = np.abs(np.random.normal(0, 1, 1000))
mu = alpha + (beta * x)
y = np.random.normal(mu, sigma)
```
